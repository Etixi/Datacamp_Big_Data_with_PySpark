```python
# Import the required functions
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.recommendation import ALS
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

# Create test and train set
(train, test) = ratings.randomSplit([0.8, 0.2], seed = 1234) #,

# Create ALS model
als = ALS(userCol="userId", itemCol="movieId", ratingCol="rating",
          nonnegative = True, implicitPrefs = False, coldStartStrategy="drop")

# Confirm that a model called "als" was created
type(als)
```

####**Dites à Spark comment régler votre modèle ALS**
+ Nous devons maintenant créer un ParamGrid pour indiquer à Spark quels hyperparamètres nous voulons qu'il règle, comment les régler, puis créer un évaluateur afin que Spark puisse savoir comment mesurer les performances de l'algorithme.

####**Instructions**

+ Importez RegressionEvaluator depuis pyspark.ml.evaluation et ParamGridBuilder et CrossValidator depuis pyspark.ml.tuning.

+ Créez un ParamGrid appelé param_grid à l'aide du ParamGridBuilder fourni. Appelez la méthode .addGrid() sur chaque hyperparamètre en fournissant le nom du modèle et le nom de chaque hyperparamètre (ex : .addGrid(als.rank, []).

+ Faites cela pour les hyperparamètres rank, maxIter et regParam. Fournissez également les listes respectives de valeurs d'hyperparamètres que Spark doit essayer, comme indiqué ici :

      rang : [10, 50, 100, 150]
      maxItère : [5, 50, 100, 200]
      regParam : [.01, .05, .1, .15]

+ Créez un RegressionEvaluator appelé évaluateur. Définissez le metricName sur "rmse", définissez le labelCol sur "rating" et indiquez à Spark que lorsqu'il génère des prédictions, il doit appeler le predictCol "prédiction".


+ Exécutez len(param_grid) pour confirmer que param_grid a été créé et pour confirmer que le bon nombre de combinaisons d'hyperparamètres sera testé.

+ Il doit être égal au nombre de valeurs de classement * au nombre de valeurs maxIter * au nombre de valeurs regParam dans ParamGridBuilder.

```python
# Import the requisite items
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator

# Add hyperparameters and their respective values to param_grid
param_grid = ParamGridBuilder() \
            .addGrid(als.rank, [10, 50, 100, 150]) \
            .addGrid(als.maxIter, [5, 50, 100, 200]) \
            .addGrid(als.regParam, [.01, .05, .1, .15]) \
            .build()

# Define evaluator as RMSE and print length of evaluator
evaluator = RegressionEvaluator(metricName="rmse", labelCol="rating", predictionCol="prediction")
print ("Num models to be tested: ", len(param_grid))
```

####**Construire des modèles implicites**

####**Instructions**
+ Une liste vide appelée model_list est fournie ici.

+ La boucle for créera un modèle pour chaque combinaison d'hyperparamètres qui lui est donnée et le placera dans model_list.


+ Complétez la boucle for en référençant le contenu de chaque liste où r représente les éléments de la liste des rangs, mi représente les éléments de la liste maxIters, rp représente les éléments de la liste regParams et a représente les éléments de la liste alphas.

+ Vous pouvez exécuter la validation supplémentaire fournie pour être sûr.

```python
# For loop will automatically create and store ALS models
for r in ranks:
    for mi in maxIters:
        for rp in regParams:
            for a in alphas:
                model_list.append(ALS(userCol= "userId", itemCol= "songId", ratingCol= "num_plays", rank = r, maxIter = mi, regParam = rp, alpha = a, coldStartStrategy="drop", nonnegative = True, implicitPrefs = True))

# Print the model list, and the length of model_list
print (model_list, "Length of model_list: ", len(model_list))

# Validate
len(model_list) == (len(ranks)*len(maxIters)*len(regParams)*len(alphas))
```

+ Depuis les évaluations implicites les modèles utilisent des métriques d’erreur personnalisées comme ROEM et non RMSE, le déséquilibre de classe ne pose pas vraiment de problème comme il pourrait le faire dans les problèmes de classification.

+ La LSA peut toujours générer des recommandations significatives à partir de ce type de données, mais il existe des stratégies qui peuvent être prises avec les données pour essayer d’améliorer les recommandations.

####**4. Pondération des articles**

+ Par exemple, plutôt que de traiter les films inédits uniquement comme des 0, vous pouvez les pondérer plus haut si plus de gens les ont vus.

+ Cela suppose que si beaucoup de gens ont vu un film, il doit être un très bon film et doit donc être traité avec un peu plus de poids, et vice versa.

+ C’est ce qu’on appelle la pondération des articles.

####**5. Pondération des articles et pondération des utilisateurs**

+ De même, vous pouvez pondérer les films en fonction du comportement individuel de l’utilisateur.

+ Par exemple, si un utilisateur a vu beaucoup de films, vous pouvez pondérer ses films inédits plus bas en supposant que si un utilisateur a vu beaucoup de films, il sait ce qu’il aime et a délibérément choisi de NE PAS regarder les films qu’il n’a pas vus et donc ces films méritent une pondération inférieure.
