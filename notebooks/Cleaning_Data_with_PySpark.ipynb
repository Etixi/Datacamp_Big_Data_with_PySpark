####**2. Qu’est-ce que le nettoyage des données?**

+ Dans ce cours, nous définirons le « nettoyage des données » comme la préparation de données brutes à utiliser dans les pipelines de traitement.

+ Nous discuterons de ce qu’est un pipeline plus tard, mais pour l’instant, il suffit de dire que le nettoyage des données est une partie nécessaire de tout système de données de production.

+ Si vos données ne sont pas « propres », elles ne sont pas fiables et pourraient causer des problèmes plus tard.

+ De nombreuses tâches pourraient relever du nettoyage des données.

+ Quelques-uns d’entre eux incluent le reformatage ou le remplacement de texte; effectuer des calculs basés sur les données; et l’élimination des déchets ou des données incomplètes.

####**3. Pourquoi effectuer un nettoyage des données avec Spark ?**

+ La plupart des systèmes de nettoyage de données ont deux gros problèmes :
  + optimiser les performances et organiser le flux de données.
  
+ Un langage de programmation typique (tel que Perl, C++ ou même SQL standard) peut être capable de nettoyer les données lorsque vous avez de petites quantités de données.

+ Mais considérez ce qui se passe lorsque vous avez des millions, voire des milliards de données.

+ Ces langues ne seraient pas en mesure de traiter cette quantité d’informations en temps opportun.

+ Spark vous permet d’adapter votre capacité de traitement des données à l’évolution de vos besoins.

+ Au-delà des problèmes de performances, le traitement de grandes quantités de données nécessite un processus ou un pipeline d’étapes.

+ Spark permet de gérer de nombreuses tâches complexes dans un cadre unique.

####**4. Exemple de nettoyage de données**

+ Voici un exemple de nettoyage d’un petit ensemble de données.

+ On nous donne une table de noms, d’âge en années et une ville.

+ Nous voulons également supprimer toutes les lignes où les données sortent de l’ordinaire.

+ À l’aide des transformations Spark, nous pouvons créer un DataFrame avec ces propriétés et poursuivre le traitement par la suite.

####**5. Schémas Spark**
+ L’une des principales fonctions du nettoyage des données est de vérifier que toutes les données sont au format attendu.

+ Spark offre une capacité intégrée pour valider des jeux de données avec des schémas.

+ Vous avez peut-être déjà utilisé des schémas avec des bases de données ou XML ; Spark est similaire.

+ Un schéma définit et valide le nombre et les types de colonnes pour un DataFrame donné.

+ Un schéma peut contenir de nombreux types de champs différents :
  + entiers, flottants, dates, chaînes et même des tableaux ou des structures de mappage.
  
+ Un schéma défini permet à Spark de filtrer les données qui ne sont pas conformes pendant la lecture, garantissant ainsi l’exactitude attendue.

+ En outre, les schémas présentent également des avantages en termes de performances.

+ Normalement, une importation de données essaiera de déduire un schéma lors de la lecture - cela nécessite de lire les données deux fois. La définition d’un schéma limite cela à une seule opération de lecture.

####**6. Exemple de schéma Spark**

+ Chaque StructField se compose d’un nom de champ, dataType, et indique si les données peuvent être null.

+ Une fois notre schéma défini, nous pouvons l’ajouter à notre appel spark.read.format.load et le traiter par rapport à nos données. La méthode load() prend deux arguments : le nom de fichier et un schéma. C’est ici que nous appliquons notre schéma aux données en cours de chargement.

###**Immuabilité et traitement paresseux**

####**1. Immuabilité et traitement paresseux**
+ Bienvenue! Nous avons eu une discussion rapide sur le nettoyage des données, les types de données et les schémas.

+ Passons à d’autres concepts Spark - Immuabilité et Traitement paresseux.

####**2. Examen des variables**
+ Normalement, en Python, et dans la plupart des autres langages, les variables sont entièrement mutables.

+ Les valeurs peuvent être modifiées à tout moment, en supposant que la portée de la variable est valide.

+ Bien que très flexible, cela pose des problèmes chaque fois que plusieurs composants simultanés tentent de modifier les mêmes données.

+ La plupart des langages contournent ces problèmes en utilisant des constructions telles que les mutex, les sémaphores, etc.

+ Cela peut ajouter de la complexité, en particulier avec des programmes non triviaux.

####**3. Immuabilité**
+ Contrairement aux variables Python typiques, les trames de données Spark sont immuables.

+ Bien qu’elle ne soit pas strictement requise, l’immuabilité est souvent une composante de la programmation fonctionnelle.

+ Nous n’entrerons pas dans tout ce que cela implique ici, mais comprenez que Spark est conçu pour utiliser des objets immuables.

+ En pratique, cela signifie que les trames de données Spark sont définies une seule fois et ne sont pas modifiables après l’initialisation.

+ Si le nom de la variable est réutilisé, les données d’origine sont supprimées (en supposant qu’elles ne sont pas utilisées ailleurs) et le nom de la variable est réaffecté aux nouvelles données.

+ Bien que cela semble inefficace, cela permet en fait à Spark de partager des données entre tous les composants du cluster.

+ Il peut le faire sans se soucier des objets de données simultanés.

####**4. Exemple d’immuabilité**

+ Il s’agit d’un exemple rapide de l’immuabilité des trames de données dans Spark.

+ Ce n’est pas grave si vous ne comprenez pas le code réel, cet exemple concerne davantage les concepts de ce qui se passe.

+ Tout d’abord, nous créons un cadre de données à partir d’un fichier CSV appelé voterdata.csv.

+ Cela crée une nouvelle définition de bloc de données et l’affecte au nom de variable voter_df.

+ Une fois créé, nous voulons faire deux autres opérations.

+ La première consiste à créer une colonne d’année complète en utilisant une année à 2 chiffres présente dans l’ensemble de données et en ajoutant 2000 à chaque entrée.

+ Cela ne change pas du tout le cadre de données.

+ Il copie la définition d’origine, ajoute la transformation et l’affecte au nom de la variable voter_df.

+ Notre deuxième opération est similaire - nous voulons maintenant supprimer la colonne de l’année d’origine du cadre de données.

+ Encore une fois, cela copie la définition, ajoute une transformation et réaffecte le nom de la variable à ce nouvel objet.

+ Les objets originaux sont détruits.

+ Veuillez noter que la colonne d’année d’origine a maintenant définitivement disparu de cette instance, mais pas des données sous-jacentes (c’est-à-dire que vous pouvez simplement la recharger dans une nouvelle trame de données si vous le souhaitez).

####**5. Traitement paresseux**

+ Vous vous demandez peut-être comment Spark fait cela si rapidement, en particulier sur les grands ensembles de données.

+ Spark peut le faire à cause de quelque chose appelé traitement paresseux.

+ Le traitement paresseux dans Spark est l’idée que très peu de choses se produisent réellement jusqu’à ce qu’une action soit effectuée.

+ Dans notre exemple précédent, nous lisions un fichier CSV, ajoutions une nouvelle colonne et en supprimions une autre.

+ L’astuce est qu’aucune donnée n’a été lue / ajoutée / modifiée, nous avons seulement mis à jour les instructions (aka, Transformations) pour ce que nous voulions que Spark fasse.

+ Cette fonctionnalité permet à Spark d’effectuer l’ensemble d’opérations le plus efficace pour obtenir le résultat souhaité.

+ L’exemple de code est identique à la diapositive précédente, mais avec l’appel de méthode count() ajouté.

+ Ceci est classé comme une action dans Spark et traitera toutes les opérations de transformation.

###**Comprendre le parquet**

####**1. Comprendre le parquet**
+ Bienvenue! Comme nous l’avons vu, Spark peut lire dans le texte et les fichiers CSV.

+ Bien que cela nous donne accès à de nombreuses sources de données, ce n’est pas toujours le format le plus pratique pour travailler.

+ Jetons un coup d’œil à quelques problèmes avec les fichiers CSV.

####**2. Difficultés avec les fichiers CSV**
+ Certains problèmes courants avec les fichiers CSV incluent: Le schéma n’est pas défini :
  + il n’y a pas de types de données inclus, ni de noms de colonnes (au-delà d’une ligne d’en-tête).
  
+ L’utilisation de contenu contenant une virgule (ou un autre délimiteur) nécessite un échappement.

+ L’utilisation du caractère d’échappement dans le contenu nécessite un échappement encore plus poussé. Les formats d’encodage disponibles sont limités en fonction de la langue utilisée.

####**3. Fichiers Spark et CSV**

+ En plus des problèmes avec les fichiers CSV en général, Spark a des problèmes spécifiques lors du traitement des données CSV.

+ Les fichiers CSV sont assez lents à importer et à analyser.

+ Les fichiers ne peuvent pas être partagés entre les travailleurs pendant le processus d’importation.

+ Si aucun schéma n’est défini, toutes les données doivent être lues avant qu’un schéma puisse être déduit.

+ Spark a une fonctionnalité connue sous le nom de prédicat pushdown.

+ Fondamentalement, c’est l’idée d’ordonner les tâches pour faire le moins de travail.

+ Le filtrage des données avant le traitement est l’une des principales optimisations du pushdown des prédicats.

+ Cela réduit considérablement la quantité d’informations qui doivent être traitées dans de grands ensembles de données.

+ Malheureusement, vous ne pouvez pas filtrer les données CSV via le menu pushdown des prédicats.

+ Enfin, les processus Spark sont souvent en plusieurs étapes et peuvent utiliser une représentation de fichier intermédiaire.

+ Ces représentations permettent d’utiliser les données ultérieurement sans régénérer les données de la source.

+ L’utilisation de CSV nécessiterait plutôt un travail supplémentaire important pour définir les schémas, les formats de codage, etc.

####**4. Le format parquet**
+ Parquet est un format de données en colonnes compressées développé pour être utilisé dans n’importe quel système basé sur Hadoop.

+ **Cela inclut Spark, Hadoop, Apache Impala, etc**.

+ Le format Parquet est structuré avec des données accessibles en morceaux, permettant des opérations de lecture / écriture efficaces sans traiter l’intégralité du fichier.

+ Ce format structuré prend en charge la fonctionnalité pushdown de prédicats de Spark, offrant une amélioration significative des performances.

+ Enfin, les fichiers Parquet incluent automatiquement les informations de schéma et gèrent l’encodage des données.

+ Ceci est parfait pour la représentation intermédiaire ou sur disque des données traitées.

+ Notez que les fichiers Parquet sont un format de fichier binaire et ne peuvent être utilisés qu’avec les outils appropriés.

+ Ceci est en contraste avec les fichiers CSV qui peuvent être édités avec n’importe quel éditeur de texte.

####**5. Travailler avec Parquet**

+ L’interaction avec les fichiers Parquet est très simple.

+ Pour lire un fichier parquet dans un Data Frame, vous avez deux options.

+ La première consiste à utiliser la méthode **'spark.read.format'** que nous avons vue précédemment.

+ Le Data Frame, **df=spark.read.format('parquet').load('filename.parquet')** La deuxième option est la version raccourci :
  + Le cadre de données, **df=spark.read.parquet('filename.parquet')** En règle générale, la version raccourcie est la plus facile à utiliser, mais vous pouvez les utiliser de manière interchangeable.
  + L’écriture de fichiers parquet est similaire, en utilisant soit: **df.write.format('parquet').save('filename.parquet')** ou **df.write.parquet('filename.parquet')** Les versions longues de chaque option permettent des indicateurs d’option supplémentaires, par exemple lors de l’écrasement d’un fichier parquet existant.

####**6. Parquet et SQL**

+ Les fichiers parquet ont diverses utilisations dans Spark.

+ Nous avons discuté de leur utilisation comme format de données intermédiaire, mais ils sont également parfaits pour effectuer des opérations SQL.

+ Pour effectuer une requête SQL sur un fichier Parquet, nous devons d’abord créer un Data Frame via la méthode **spark.read.parquet**.

+ Une fois que nous avons le Data Frame, nous pouvons utiliser la méthode ***createOrReplaceTempView()*** pour ajouter un alias des données Parquet en tant que table SQL.

+ Enfin, nous exécutons notre requête en utilisant la syntaxe SQL normale et la méthode spark.sql.

+ Dans ce cas, nous recherchons tous les vols d’une durée inférieure à 100 minutes.

+ Parce que nous utilisons Parquet comme magasin de support, nous obtenons tous les avantages de performance dont nous avons parlé précédemment (principalement les schémas définis et l’utilisation disponible du pushdown de prédicat).

####**2. Mise à jour DataFrame**

###**Opérations de colonne DataFrame conditionnelles**

####**1. Opérations de colonne DataFrame conditionnelles**

+ *Nous avons examiné une partie de la puissance disponible lors de l’utilisation des fonctions de Spark pour filtrer et modifier nos trames de données. Passons un peu de temps avec des options plus avancées.*

####**2. Clauses conditionnelles**

+ Les transformations DataFrame que nous avons couvertes jusqu’à présent sont des transformations générales, ce qui signifie qu’elles sont appliquées quelles que soient les données.

+ Souvent, vous voudrez changer conditionnellement certains aspects du contenu.

+ Spark fournit des clauses conditionnelles intégrées qui agissent de la même manière qu’une instruction if / then / else dans un environnement de programmation traditionnel.

+ Bien qu’il soit possible d’exécuter une instruction de style if / then / else traditionnelle dans Spark, cela peut entraîner une dégradation grave des performances car chaque ligne d’un DataFrame serait évaluée indépendamment.

+ L’utilisation des conditions optimisées et intégrées atténue ce problème.

+ Les clauses conditionnelles comportent deux composants : ***.when() et le .otherwise()*** facultatif. Regardons-les plus en profondeur.

####**3. Exemple conditionnel**

+ La clause .when() est une méthode disponible dans la bibliothèque **pyspark.sql.functions** qui recherche deux composants :
  + la condition if et la condition que faire si elle donne la valeur true.
  
+ C’est ce que l’on voit le mieux à partir d’un exemple.

+ Envisagez un DataFrame avec les colonnes Nom et Âge.

+ Nous pouvons en fait ajouter un argument supplémentaire à notre méthode .select() en utilisant la clause .when().

+ Nous sélectionnons df. Nom et df.

+ Vieillir comme d’habitude. Pour le troisième argument, nous allons définir un lorsque conditionnel.

+ Si la colonne Âge est 18 ou plus, nous ajouterons la chaîne « Adulte ».

+ Si la clause ne correspond pas, rien n’est renvoyé. Notez que notre DataFrame renvoyé contient une colonne sans nom que nous n’avons pas définie à l’aide de .withColumn().
+ La fonction .select() peut créer des colonnes dynamiquement en fonction des arguments fournis. Regardons d’autres exemples.

####**4. Un autre exemple**
+ Vous pouvez chaîner plusieurs instructions when ensemble, comme une structure if / else if.

+ Dans ce cas, nous définissons deux clauses .when() et retournons Adulte ou Mineur en fonction de la colonne Age.

+ Vous pouvez enchaîner autant de clauses lorsque que nécessaire.

####**5. Autre chose**

+ En plus de .when() est la clause otherwise().

+ .otherwise() est analogue à l’instruction else. Il faut un seul argument, qui est ce qu’il faut retourner, au cas où la ou les clauses when ne seraient pas évaluées comme True.
+ Dans cet exemple, nous retournons « Adulte » lorsque la colonne Âge est 18 ou plus. Sinon, nous retournons « Mineur ».

+ Le DataFrame résultant est le même, mais la méthode est différente.

+ Bien que vous puissiez avoir plusieurs instructions .when() enchaînées, vous ne pouvez avoir qu’un seul .otherwise() par chaîne .when().

####**Partitionnement et traitement paresseux**

####**1. Partitionnement et traitement paresseux**

+ Bienvenue à notre discussion sur la modification et le nettoyage des DataFrames.

+ Nous avons discuté de diverses transformations et méthodes pour modifier nos données, mais nous n’avons pas beaucoup parlé de la façon dont Spark traite réellement les données. Regardons cela maintenant.

####**2. Partitionnement**
+ Spark divise les DataFrames en partitions ou blocs de données.

+ Ces partitions peuvent être automatiquement définies, agrandies, réduites et peuvent différer considérablement en fonction du type de cluster Spark utilisé.

+ La taille de la partition varie, mais essayez généralement de garder vos tailles de partition égales.

+ Nous aborderons plus en détail l’optimisation du partitionnement et les détails du cluster plus tard.

+ Pour l’instant, supposons que chaque partition soit gérée indépendamment.

+ Cela fait partie de ce qui fournit les niveaux de performances et la capacité de mise à l’échelle horizontale dans Spark.

+ Si un nœud Spark n’a pas besoin de rivaliser pour les ressources, ni de consulter d’autres nœuds Spark pour obtenir des réponses, il peut planifier de manière fiable le traitement pour obtenir les meilleures performances.

####**3. Traitement paresseux**
+ Dans Spark, toute opération de transformation est paresseuse ; Cela ressemble plus à une recette qu’à une commande.

+ Il définit ce qui doit être fait à un DataFrame plutôt que de le faire réellement.

+ La plupart des opérations dans Spark sont en fait des transformations, y compris **.withColumn(), .select(), .filter(), etc.**

+ L’ensemble des transformations que vous définissez n’est exécuté que lorsque vous exécutez une action Étincelle.

+ Cela inclut ***.count(), .write(), etc***.
  - tout ce qui nécessite l’exécution des transformations pour obtenir correctement une réponse.
  
+ Spark peut réorganiser les transformations pour obtenir les meilleures performances.

+ Habituellement, cela n’est pas perceptible, mais peut parfois provoquer un comportement inattendu, tel que des ID qui ne sont pas ajoutés tant que d’autres transformations ne sont pas terminées.

+ Cela ne pose pas vraiment de problème, mais les données peuvent sembler inhabituelles si vous ne savez pas à quoi vous attendre.

####**4. Ajout d’ID**
+ Les bases de données relationnelles ont tendance à avoir un champ utilisé pour identifier la ligne, que ce soit pour une référence de relation réelle ou simplement pour l’identification des données.

+ Ces ID sont généralement un entier dont la valeur augmente, qui est séquentiel et, surtout, unique.

+ Le problème avec ces identifiants est qu’ils ne sont pas de nature très parallèle.

+ Étant donné que les valeurs sont données séquentiellement, s’il y a plusieurs travailleurs, ils doivent tous se référer à une source commune pour l’entrée suivante.

+C’est correct dans un environnement de serveur unique, mais dans une plate-forme distribuée telle que Spark, cela crée des goulots d’étranglement indus. Voyons comment générer des ID dans Spark.

####**5. Augmentation monotone des identifiants**

+ Spark dispose d’une fonction intégrée appelée **monotonically_increasing_id()**, conçue pour fournir un ID entier dont la valeur augmente et qui est unique.

+ Ces ID ne sont pas nécessairement séquentiels
  - il peut y avoir des écarts, souvent assez importants, entre les valeurs.
  
+ Contrairement à un ID relationnel normal, celui de Spark est complètement parallèle
  - chaque partition est allouée jusqu’à 8 milliards d’ID qui peuvent être attribués.
+ Notez que les champs ID de l’exemple de table sont des entiers, dont la valeur augmente, mais ne sont pas séquentiels.
+ C’est un peu hors de portée, mais les ID sont un nombre de 64 bits effectivement divisé en groupes basés sur la partition Spark.
+ Chaque groupe contient 8,4 milliards d’identifiants, et il y a 2,1 milliards de groupes possibles, dont aucun ne se chevauche.

####**6. Remarques**
+ Il y a beaucoup de nuances dans la façon dont les partitions et l’augmentation monotone de l’ID fonctionnent.
+ Rappelez-vous que Spark est paresseux:
  + il aide souvent à résoudre ce qui peut arriver.
+ Les opérations sont souvent hors service, surtout s’il s’agit de jointures. Il est préférable de tester vos transformations.

###**Mise en cache**

####**1. Mise en cache**

+ Maintenant que nous avons effectué quelques tâches de nettoyage de données à l’aide de Spark, voyons comment améliorer les performances de l’exécution de ces tâches à l’aide de la mise en cache.

####**2. Qu’est-ce que la mise en cache ?**

+ La mise en cache dans Spark fait référence au stockage des résultats d’un DataFrame en mémoire ou sur le disque des nœuds de traitement d’un cluster.

+ La mise en cache améliore la vitesse des transformations ou actions ultérieures, car les données n’ont probablement plus besoin d’être extraites de la source de données d’origine.

+ L’utilisation de la mise en cache réduit l’utilisation des ressources du cluster : il est moins nécessaire d’accéder au stockage, à la mise en réseau et au processeur des nœuds Spark, car les données sont probablement déjà présentes.

####**3. Inconvénients de la mise en cache**

+ Il y a quelques inconvénients de la mise en cache que vous devez connaître.

+ Les ensembles de données très volumineux peuvent ne pas tenir dans la mémoire réservée aux DataFrames mis en cache.

+ En fonction des transformations ultérieures demandées, le cache peut ne rien faire pour améliorer les performances.

+ Si un ensemble de données ne reste pas mis en cache en mémoire, il peut être conservé sur le disque.

+ Selon la configuration de disque d’un cluster Spark, il peut ne pas s’agir d’une amélioration significative des performances.

+ Si vous lisez à partir d’une ressource réseau locale et que vos E/S de disque local sont lentes, il peut être préférable d’éviter la mise en cache des objets.

+ Enfin, la durée de vie d’un objet mis en cache n’est pas garantie. Spark gère automatiquement la régénération des DataFrames, mais cela peut entraîner des retards de traitement.

####**4. Conseils de mise en cache**

+ La mise en cache est incroyablement utile, mais uniquement si vous prévoyez d’utiliser à nouveau le DataFrame.

+ Si vous n’en avez besoin que pour une seule tâche, cela ne vaut pas la peine d’être mis en cache.

+ La meilleure façon d’évaluer les performances avec la mise en cache consiste à tester différentes configurations.

+ Essayez de mettre en cache vos DataFrames à différents moments du cycle de traitement et vérifiez si cela améliore votre temps de traitement.

+ Essayez de mettre en cache en mémoire ou un stockage NVMe / SSD rapide.

+ Bien qu’encore plus lent que la mémoire principale, le stockage SSD moderne est considérablement plus rapide que le disque en rotation.

+ Les disques durs rotatifs locaux peuvent toujours être utiles si vous traitez des DataFrames volumineux dont la génération nécessite de nombreuses étapes ou qui doivent être accessibles via Internet. Il est crucial de tester cela.

+ Si la mise en cache normale ne semble pas fonctionner, essayez de créer des représentations Parquet intermédiaires comme nous l’avons fait dans le chapitre 1.

+ Ceux-ci peuvent fournir un point de contrôle en cas d’échec d’une tâche en cours de tâche et peuvent toujours être utilisés avec la mise en cache pour améliorer encore les performances.

+ Enfin, vous pouvez arrêter manuellement la mise en cache d’un DataFrame lorsque vous en avez terminé. Cela libère des ressources de cache pour d’autres DataFrames.

####**5. Mise en œuvre de la mise en cache**

+ L’implémentation de la mise en cache dans Spark est simple.

+ La méthode principale consiste à appeler la fonction .cache() sur un objet DataFrame avant une action donnée.

+ Il ne nécessite aucun argument.

+ Un exemple est la création d’un DataFrame à partir de données CSV d’origine. Avant d’exécuter un fichier .count() sur les données, nous appelons .cache() pour indiquer à Spark de le stocker dans le cache.

+ Une autre option consiste à appeler .cache() séparément. Ici, nous créons un ID en une seule transformation.

+ Nous appelons ensuite .cache() sur le DataFrame.

+ Lorsque nous appelons l’action .show(), le DataFrame voter_df est mis en cache.

+ Si vous suivez de près, cela signifie que .cache() est une transformation Spark - rien n’est réellement mis en cache tant qu’une action n’est pas appelée.

####**6. Plus d’opérations de cache**
+ Quelques autres options sont disponibles avec la mise en cache dans Spark.

+ Pour vérifier si un DataFrame est mis en cache, utilisez la propriété booléenne .is_cached qui renvoie True (comme dans ce cas) ou False. Pour annuler la mise en cache d’un DataFrame, nous appelons .unpersist() sans argument. Cela supprime l’objet du cache.

####**1. Conseils de dimensionnement des grappes**

####**2. Options de configuration**

+ Spark dispose de nombreux paramètres de configuration disponibles contrôlant tous les aspects de l’installation.

+ Ces configurations peuvent être modifiées pour répondre au mieux aux besoins spécifiques du cluster.

+ Les configurations sont disponibles dans les fichiers de configuration, via l’interface Web Spark et via le code d’exécution.

+ Pour lire un paramètre de configuration, appelez **spark.conf.get()** avec le nom du paramètre comme argument.

+ Pour écrire un paramètre de configuration, appelez **spark.conf.set()** avec le nom du paramètre et la valeur réelle comme arguments de fonction.

####**3. Types de grappes**

+ Les déploiements Spark peuvent varier en fonction des besoins exacts des utilisateurs. Un composant important d’un déploiement est le mécanisme de gestion de cluster.

+ Les grappes d’étincelles peuvent être :
  + ***Clusters à nœud unique, déployant tous les composants sur un seul système (physique / VM / conteneur)***.
  
+ Clusters autonomes, avec des machines dédiées comme pilote et travailleurs.

+ Clusters gérés, ce qui signifie que les composants de cluster sont gérés par un gestionnaire de cluster tiers tel que **YARN, Mesos ou Kubernetes**.

+ Dans ce cours, nous utilisons un cluster à nœuds unique. Votre environnement de production peut varier énormément, mais nous aborderons les clusters autonomes au fur et à mesure que les concepts circulent dans tous les types de gestion.

####**4. Conducteur**
+ Si vous vous souvenez, il y a un pilote par cluster Spark.

+ Le conducteur est responsable de plusieurs choses, dont les suivantes :
  + Gestion de l’affectation des tâches aux différents nœuds / processus du cluster.
  
+ Le pilote surveille l’état de tous les processus et tâches et gère toutes les nouvelles tentatives.

+ Le pilote est également responsable de la consolidation des résultats des autres processus du cluster.

+ Le pilote gère tout accès aux données partagées et vérifie que chaque processus de travail dispose des ressources nécessaires (code, données, etc.).

+ Compte tenu de l’importance du pilote, il vaut souvent la peine d’augmenter les spécifications du nœud par rapport à d’autres systèmes.

+ Il est recommandé de doubler la mémoire par rapport aux autres nœuds.

+ Ceci est utile pour les tâches de surveillance des tâches et de consolidation des données.

+ Comme avec tous les systèmes Spark, un stockage local rapide est utile pour exécuter Spark dans une configuration idéale.

####**5. Travailleur**

+ Un collaborateur Spark gère les tâches en cours d’exécution assignées par le pilote et communique ces résultats au pilote.

+ Idéalement, le travailleur dispose d’une copie de tout le code, des données et de l’accès aux ressources nécessaires pour effectuer une tâche donnée.

+ Si l’un de ces éléments n’est pas disponible, le collaborateur doit faire une pause pour obtenir les ressources.

+ Lors du dimensionnement d’un cluster, il existe quelques recommandations :
  + Selon le type de tâche, un plus grand nombre de nœuds de travail est souvent préférable à celui de nœuds plus grands.
  
+ Cela peut être particulièrement évident lors des opérations d’importation et d’exportation car il y a plus de machines disponibles pour faire le travail.

+ Comme pour tout dans Spark, testez différentes configurations pour trouver le bon équilibre pour votre charge de travail.

+ En supposant un environnement cloud, 16 nœuds de travail peuvent effectuer une tâche en une heure et coûter 50 $ en ressources.

+ Une configuration de 8 travailleurs peut prendre 1,25 heure, mais ne coûte que la moitié du prix.

+ Enfin, les travailleurs peuvent utiliser un stockage local rapide (SSD / NVMe) pour la mise en cache, les fichiers intermédiaires, etc.

####**3. Qu’est-ce que le brassage?**
+ Spark distribue les données entre les différents nœuds du cluster.

+ Un effet secondaire de ceci est ce que l’on appelle le brassage.

+ Le brassage est le déplacement de fragments de données vers divers travailleurs selon les besoins pour effectuer certaines tâches.

+ Le brassage est utile et cache la complexité globale à l’utilisateur (l’utilisateur n’a pas besoin de savoir quels nœuds ont quelles données).

+ Cela étant dit, il peut être lent d’effectuer les transferts nécessaires, surtout si quelques nœuds nécessitent toutes les données.

+ La lecture aléatoire réduit le débit global du cluster, car les travailleurs doivent attendre le transfert des données.

+ Cela limite le nombre de travailleurs disponibles pour les tâches restantes dans le système.

+ Le brassage est souvent un élément nécessaire, mais il est utile d’essayer de le minimiser autant que possible.

####**4. Comment limiter le brassage ?**

+ Il peut être délicat de supprimer complètement les opérations de brassage, mais il y a quelques choses qui peuvent le limiter.

+ La fonction DataFrame .repartition() prend un seul argument, le nombre de partitions demandées.

+ Nous l’avons utilisé dans un chapitre précédent pour illustrer l’effet des partitions avec la fonction monotonically_increasing_id().

+ Le repartitionnement nécessite un mélange complet de données entre les nœuds et les processus et est assez coûteux. Si vous devez réduire le nombre de partitions, utilisez plutôt la fonction .coalesce().

+ Il prend un nombre de partitions plus petit que l’actuel et consolide les données sans nécessiter un mélange complet des données.

+ Remarque: appeler .coalesce() avec un plus grand nombre de partitions ne fait rien.


+ La fonction .join() est une excellente utilisation de Spark et fournit beaucoup de puissance.

+ L’appel de .join() sans discernement peut souvent provoquer des opérations de mélange, entraînant une augmentation de la charge du cluster et des temps de traitement plus lents.

+ Pour éviter certaines opérations de lecture aléatoire lors de la jonction de Spark DataFrames, vous pouvez utiliser la fonction .broadcast().

+ Nous en reparlerons dans un instant.

+ Enfin, une note importante sur les opérations de nettoyage des données est de se rappeler d’optimiser pour ce qui compte.

+ La vitesse de votre code initial peut être parfaitement acceptable et le temps peut être mieux dépensé ailleurs.

####**5. Diffusion**
+ La diffusion dans Spark est une méthode permettant de fournir une copie d’un objet à chaque travailleur.

+ Lorsque chaque travailleur possède sa propre copie des données, il y a moins besoin de communication entre les nœuds.

+ Cela limite les mélanges de données et il est plus probable qu’un nœud accomplisse les tâches de manière indépendante.

+ L’utilisation de la diffusion peut considérablement accélérer les opérations .join(), en particulier si l’une des DataFrames jointes est beaucoup plus petite que l’autre.

+ Pour implémenter la diffusion, vous devez importer la fonction de diffusion à partir de pyspark.sql.functions.

+ Spark optimisera souvent cela pour vous, mais comme d’habitude, exécutez des tests dans votre environnement pour obtenir les meilleures performances.

###**Introduction aux pipelines de données**

####**1. Introduction aux pipelines de données**

####**2. Qu’est-ce qu’un pipeline de données ?**

+ Les pipelines de données sont simplement l’ensemble des étapes nécessaires pour passer d’une ou plusieurs sources de données d’entrée et les convertir en sortie souhaitée.

+ Un pipeline de données peut être composé d’un nombre illimité d’étapes ou de composants et peut couvrir de nombreux systèmes.

+ Pour nos besoins, nous allons configurer un pipeline de données dans Spark, mais nous sommes conscients qu’un pipeline de données de production complet communiquera probablement avec de nombreux systèmes.

####**3. À quoi ressemble un pipeline de données ?**

+ Tout comme Spark en général, un pipeline de données se compose généralement d’entrées, de transformations et de sorties de ces étapes.

+ En outre, il y a souvent des étapes de validation et d’analyse avant la livraison des données à l’utilisateur suivant.

+ Une entrée peut être l’un des types de données que nous avons examinés jusqu’à présent, y compris CSV, JSON, texte, etc.

+ Il peut provenir du système de fichiers local ou de services Web, d’API, de bases de données, etc. L’idée de base est de lire les données dans un DataFrame comme nous l’avons fait précédemment.

+ Une fois que nous avons les données dans un DataFrame, nous devons les transformer d’une manière ou d’une autre.

+ Vous avez effectué les étapes individuelles plusieurs fois au cours de ce cours
  - ajout de colonnes, filtrage de lignes, exécution de calculs selon les besoins.
  
+ Dans nos exemples précédents, nous n’avons effectué qu’une ou deux de ces étapes à la fois, mais un pipeline peut comprendre autant d’étapes que nécessaire afin que nous puissions formater les données dans la sortie souhaitée.

+ Une fois que nous avons défini nos transformations, nous devons sortir les données sous une forme utilisable.

+ Vous avez déjà écrit des fichiers au format CSV ou Parquet, mais il peut inclure plusieurs copies avec différents formats, ou plutôt écrire la sortie dans une base de données, un service Web, etc. Les deux dernières étapes varient grandement en fonction de vos besoins.

+ Nous discuterons de la validation dans une leçon ultérieure, mais l’idée est d’exécuter une forme de test sur les données pour vérifier qu’elles sont comme prévu.

+ L’analyse est souvent la dernière étape avant de transmettre les données à l’utilisateur suivant.

+ Cela peut inclure des éléments tels que le nombre de lignes, des calculs spécifiques ou à peu près tout ce qui facilite la consommation du jeu de données par l’utilisateur.

####**4. Détails du pipeline**

+ Il est important de noter que dans Spark, un pipeline de données n’est pas un objet formellement défini, mais plutôt un concept.

```python
def retriever(cols, colcount):
  # Return a list of dog data
  return cols[4:colcount]

# Define the method as a UDF
udfRetriever = F.udf(retriever, ArrayType(StringType()))

# Create a new column using your UDF
split_df = split_df.withColumn('dog_list', udfRetriever(split_df.split_cols, split_df.colcount))

# Remove the original column, split_cols, and the colcount
split_df = split_df.drop('_c0').drop('split_cols').drop('colcount')
```

+ Les jointures sont également relativement rapides, en particulier par rapport à la validation de lignes individuelles par rapport à une longue liste d’entrées.

####**Pourcentage de pixels de chien**
+ La dernière tâche d'analyse des données d'annotation de chien consiste à déterminer le pourcentage de pixels dans chaque image qui représente un chien (ou des chiens).

+ Pour calculer le pourcentage de pixels, calculez d'abord le nombre total de pixels représentant chaque chien, puis additionnez-les pour l'image.

+ Vous pouvez calculer la boîte englobante avec la formule :

  + $(Xend - Xstart) * (Yend - Ystart)$

+ REMARQUE :
  + vous pouvez ignorer la possibilité de chevauchement des cadres de délimitation dans ce cas.

+ *Pour le pourcentage, calculez le nombre total de pixels "chien" divisé par la taille totale de l'image, multiplié par 100.

####**Instructions**
+ Définissez une fonction Python pour prendre une liste de tuples (les objets chien) et calculer le nombre total de pixels "chien" par image.
+ Créez un UDF de la fonction et utilisez-le pour créer une nouvelle colonne appelée 'dog_pixels' sur le DataFrame.
+ Créez une autre colonne, 'dog_percent', représentant le pourcentage de 'dog_pixels' dans l'image.

+ Assurez-vous qu'il se situe entre 0 et 100 %.

+ Utilisez uniquement le nom de chaîne de la colonne (c'est-à-dire, "columnname" plutôt que df.columnname).

+ Utilisez une chaîne de style SQL pour cela (par exemple, 'columnname> ____')

```python
# Define a UDF to determine the number of pixels per image
def dogPixelCount(doglist):
  totalpixels = 0
  for dog in doglist:
    totalpixels += (dog[3] - dog[1]) * (dog[4] - dog[2])
  return totalpixels

# Define a UDF for the pixel count
udfDogPixelCount = F.udf(dogPixelCount, IntegerType())
joined_df = joined_df.withColumn('dog_pixels', udfDogPixelCount('dogs'))

# Create a column representing the percentage of pixels
joined_df = joined_df.withColumn('dog_percent', (joined_df.dog_pixels / (joined_df.width * joined_df.height)) * 100)

# Show the first 10 annotations with more than 60% dog
joined_df.where('dog_percent > 60').show(10)
```
