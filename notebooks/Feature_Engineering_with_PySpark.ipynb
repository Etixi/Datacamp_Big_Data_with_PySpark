####**5. Formats de données: Parquet**


+ Pour ce cours, nous utiliserons un fichier Parquet.

+ Comme la plupart des données dans Hadoop, la plate-forme sur laquelle Spark fonctionne, il s’agit d’un format d’écriture une fois, lu plusieurs fois.

+ Les données Parquet sont en colonnes, ce qui signifie qu’elles sont organisées par colonnes, une caractéristique importante pour les ensembles de données volumineux car il est extrêmement rapide de lire UNIQUEMENT les données dont vous avez besoin.

+ Les CSV, en revanche, doivent lire et analyser l’ensemble des données pour lire un seul champ.

+ Une autre différence est que les champs Parquet sont définis et tapés, ce qui évite aux utilisateurs de définir des types de données, tels que des dates, des booléens ou des chaînes.

+ Pour cette raison, le parquet est relativement lent à écrire.

+ Comme il n’est pas délimité par des caractères, il est moins susceptible d’être mal lu si ces caractères existent dans les données.

+ Ce ne sont là que quelques avantages qui poussent l’industrie à adopter rapidement le parquet.

+ À l'aide de **num_records**, créez une vérification pour voir si la trame de données d'entrée df a le même montant avec **count()**.

+ Si les deux renvoient Vrai, imprimez Validation réussie

```python
def check_load(df, num_records, num_columns):
  # Takes a dataframe and compares record and column counts to input
  # Message to return if the critera below aren't met
  message = 'Validation Failed'
  # Check number of records
  if num_records == df.count():
    # Check number of columns
    if num_columns == len(df.columns):
      # Success message
      message = 'Validation Passed'
  return message
```

####**Mise à l'échelle de vos scalers**

+ Dans l'exercice précédent, nous avons mis à l'échelle minmax une seule variable.

+ Supposons que vous ayez BEAUCOUP de variables à mettre à l'échelle, vous ne voulez pas que des centaines de lignes codent pour chacune.

####**Instructions**

+ Définissez une fonction appelée min_max_scaler qui prend les paramètres df une trame de données et cols_to_scale la liste des colonnes à mettre à l'échelle.
+ Utilisez une boucle for pour parcourir chaque colonne de la liste et les mettre à l'échelle minmax.
+ Renvoie le dataframe df avec les nouvelles colonnes ajoutées.
+ Appliquez la fonction min_max_scaler() sur df et la liste des colonnes cols_to_scale.

```python
def min_max_scaler(df, cols_to_scale):
  # Takes a dataframe and list of columns to minmax scale. Returns a dataframe.
  for col in cols_to_scale:
    # Define min and max values and collect them
    max_days = df.agg({col: 'max'}).collect()[0][0]
    min_days = df.agg({col: 'min'}).collect()[0][0]
    new_column_name = 'scaled_' + col
    # Create a new column based off the scaled data
    df = df.withColumn(new_column_name,
                      (df[col] - min_days) / (max_days - min_days))
  return df
```

+ Pour corriger le biais négatif (données principalement à droite), vous devez effectuer une étape supplémentaire appelée « réflexion » avant de pouvoir appliquer l'inverse de , écrit sous la forme ($\frac{1}{log}$) pour rendre les données plus attrayantes comme une distribution normale.

+ La réflexion des données utilise la formule suivante pour refléter chaque valeur : $(x_{\text{max}} +1) – x$.

```python
from pyspark.sql.functions import log

# Compute the skewness
print(df.agg({'YEARBUILT': 'skewness'}).collect())

# Calculate the max year
max_year = df.agg({'YEARBUILT': 'max'}).collect()[0][0]

# Create a new column of reflected data
df = df.withColumn('Reflect_YearBuilt', (max_year + 1) - df['YEARBUILT'])

# Create a new column based reflected data
df = df.withColumn('adj_yearbuilt', 1 / log(df['Reflect_YearBuilt']))
```

####**3. Types de manquants**

+ L’option Complètement manquante au hasard se produit lorsque les données sont manquantes sans modèle.

####**Calculer les pourcentages manquants**

+ L'automatisation est l'avenir de la science des données.

+ Apprendre à automatiser une partie de la préparation de vos données porte ses fruits.

+ Dans cet exercice, nous automatiserons la suppression de colonnes s'il manque des données au-delà d'un seuil spécifique.

####**Instructions**

+ Définissez une fonction column_dropper() qui prend les paramètres df une trame de données et un seuil flottant entre 0 et 1.


+ Calculez le pourcentage de valeurs manquantes en utilisant where(), isNull() et count()


+ Vérifiez si le pourcentage de manquants est supérieur au seuil, si c'est le cas, supprimez la colonne à l'aide de drop ()


+ Exécutez column_dropper() sur df avec le seuil défini sur 0,6

```python
def column_dropper(df, threshold):
  # Takes a dataframe and threshold for missing values. Returns a dataframe.
  total_records = df.count()
  for col in df.columns:
    # Calculate the percentage of missing values
    missing = df.where(df[col].isNull()).count()
    missing_percent = missing / total_records
    # Drop column if percent of missing is more than threshold
    if missing_percent > threshold:
      df = df.drop(col)
  return df
```

```python
# Drop columns that are more than 60% missing
df = column_dropper(df, .6)
```

```python
from pyspark.sql.functions import lag, datediff, to_date
from pyspark.sql.window import Window

# Cast data type
mort_df = mort_df.withColumn('DATE', to_date('DATE'))

# Create window
w = Window().orderBy(mort_df['DATE'])
# Create lag column
mort_df = mort_df.withColumn('DATE-1', lag('DATE', count=1).over(w))

# Calculate difference between date columns
mort_df = mort_df.withColumn('Days_Between_Report', datediff('DATE', 'DATE-1'))
# Print results
mort_df.select('Days_Between_Report').distinct().show()
```

```python
from datetime import timedelta
```

```python
def train_test_split_date(df, split_col, test_days=45):
  """Calculate the date to split test and training sets"""
  # Find how many days our data spans
  max_date = df.agg({split_col: 'max'}).collect()[0][0]
  min_date = df.agg({split_col: 'min'}).collect()[0][0]
  # Subtract an integer number of days from the last date in dataset
  split_date = max_date - timedelta(days=test_days)
  return split_date
```
