###**Qu'est-ce que Spark, d'ailleurs ?**

+ **Spark** est une plateforme pour le calcul en **grappe**. Spark vous permet de répartir les données et les calculs sur **des clusters à plusieurs nœuds** (pensez à chaque nœud comme à un ordinateur distinct).

+ **La division de vos données** permet de travailler plus facilement avec de très grands ensembles de données, car chaque nœud ne travaille que sur une petite quantité de données.

+ Comme chaque **nœud** travaille sur son propre sous-ensemble de données, il effectue également une partie des calculs nécessaires, de sorte que le traitement des données et les calculs sont effectués en parallèle sur les nœuds du cluster. C'est un fait que le calcul parallèle peut rendre certains types de tâches de programmation beaucoup plus rapides.

+ Cependant, une plus grande puissance de calcul s'accompagne d'une plus grande complexité.

+ Décider si Spark est la meilleure solution pour votre problème demande une certaine expérience, mais vous pouvez considérer des questions telles que :

  + Mes données sont-elles trop volumineuses pour être traitées sur une seule machine ?

  + Mes calculs peuvent-ils être facilement parallélisés ?

**Utiliser Spark en Python**

+ La première étape de l'utilisation de Spark consiste à se connecter à un cluster.

+ En pratique, le cluster sera hébergé sur une machine distante qui est connectée à tous les autres nœuds.

+ Il y aura un ordinateur, appelé **le maître**, qui gère la répartition des données et des calculs.

+ Le maître est connecté au reste des ordinateurs du cluster, qui sont appelés travailleurs.

+ Le maître envoie aux travailleurs les données et les calculs à exécuter, et ceux-ci renvoient leurs résultats au maître.

+ Lorsque vous commencez à utiliser Spark, il est plus simple d'exécuter un cluster localement. Ainsi, pour ce cours, au lieu de se connecter à un autre ordinateur, tous les calculs seront exécutés sur les serveurs de DataCamp dans un cluster simulé.

+ **La création de la connexio**n est aussi simple que la création d'une instance de la **classe SparkContext**. Le constructeur de la classe prend quelques arguments optionnels qui vous permettent de spécifier les attributs du cluster auquel vous vous connectez.

+ **Examinons comment effectuer des opérations par colonne. Dans Spark, vous pouvez le faire en utilisant la méthode .withColumn(), qui prend deux arguments. Premièrement, une chaîne avec le nom de votre nouvelle colonne, et deuxièmement la nouvelle colonne elle-même.**

+ **La nouvelle colonne doit être un objet de la classe Column. Créer l'un d'entre eux est aussi simple que d'extraire une colonne de votre DataFrame à l'aide de df.colName.**

+ **Ainsi, toutes ces méthodes renvoient un nouveau DataFrame. Pour écraser le DataFrame d'origine, vous devez réaffecter le DataFrame renvoyé en utilisant la méthode suivante :**

      df = df.withColumn("newCol", df.oldCol + 1)

+ ***Le code ci-dessus crée un DataFrame avec les mêmes colonnes que df plus une nouvelle colonne, newCol, où chaque entrée est égale à l'entrée correspondante de oldCol, plus un.***

+ **Pour écraser une colonne existante, passez simplement le nom de la colonne comme premier argument !**

####**SQL en bref**

+ ***Au fur et à mesure que vous avancez, cela vous aidera à avoir une compréhension de base de SQL. Un regard plus approfondi peut être trouvé [ici](https://app.datacamp.com/learn/courses/introduction-to-sql).***

+ ***Une requête SQL renvoie une table dérivée d'une ou plusieurs tables contenues dans une base de données.**

+ **Chaque requête SQL est composée de commandes qui indiquent à la base de données ce que vous voulez faire avec les données. Les deux commandes que chaque requête doit contenir sont SELECT et FROM.**

+ La commande SELECT est suivie des colonnes souhaitées dans la table résultante.

+ La commande FROM est suivie du nom de la table qui contient ces colonnes. La requête SQL minimale est :

          SELECT * FROM my_table ;

####**Validation croisée**

+ ***Dans les prochains exercices, vous ajusterez votre modèle de régression logistique à l'aide d'une procédure appelée validation croisée k-fold. Il s'agit d'une méthode d'estimation des performances du modèle sur des données invisibles (comme votre DataFrame de test).***

+ ***Cela fonctionne en divisant les données de formation en quelques partitions différentes. Le nombre exact dépend de vous, mais dans ce cours, vous utiliserez la valeur par défaut de trois de PySpark. Une fois les données découpées, l'une des partitions est mise de côté et le modèle est ajusté aux autres. Ensuite, l'erreur est mesurée par rapport à la partition retenue. Ceci est répété pour chacune des partitions, de sorte que chaque bloc de données est retenu et utilisé comme ensemble de test exactement une fois. Ensuite, l'erreur sur chacune des partitions est moyennée. C'est ce qu'on appelle l'erreur de validation croisée du modèle, et c'est une bonne estimation de l'erreur réelle sur les données retenues.***

```python
from pyspark.ml.feature import StringIndexer, OneHotEncoder
from pyspark.ml.feature import VectorAssembler
from pyspark.ml import Pipeline

flights = (spark.read.format("csv")
  .option("inferSchema", "true")
  .option("header", "true")
  .load("./dataset/flights_small.csv"))
flights.createOrReplaceTempView("flights")

planes = (spark.read.format("csv")
  .option("inferSchema", "true")
  .option("header", "true")
  .load('./dataset/planes.csv'))
planes.createOrReplaceTempView("planes")
```

```python
# Rename year column
planes = planes.withColumnRenamed('year', 'plane_year')

# Join the DataFrame
model_data = flights.join(planes, on='tailnum', how='leftouter')

# Cast the columns to integers
model_data = model_data.withColumn("arr_delay", model_data.arr_delay.cast('integer'))
model_data = model_data.withColumn('air_time', model_data.air_time.cast('integer'))
model_data = model_data.withColumn('month', model_data.month.cast('integer'))
model_data = model_data.withColumn('plane_year', model_data.plane_year.cast('integer'))

# Create the column plane_age
model_data = model_data.withColumn('plane_age', model_data.year - model_data.plane_year)

# Create is_late
model_data = model_data.withColumn('is_late', model_data.arr_delay > 0)

# Convert to an integer
model_data = model_data.withColumn('label', model_data.is_late.cast('integer'))

# Remove missing values
model_data = model_data.filter('arr_delay is not NULL and dep_delay is not NULL and \
                               air_time is not NULL and plane_year is not NULL')

# Create StringIndexer
carr_indexer = StringIndexer(inputCol='carrier', outputCol='carrier_index')

# Create a OneHotEncoder
carr_encoder = OneHotEncoder(inputCol='carrier_index', outputCol='carrier_fact')

# Create a StringIndexer
dest_indexer = StringIndexer(inputCol='dest', outputCol='dest_index')

# Create a OneHotEncoder
dest_encoder = OneHotEncoder(inputCol='dest_index', outputCol='dest_fact')

# Make a VectorAssembler
vec_assembler = VectorAssembler(inputCols=['month', 'air_time', 'carrier_fact', 'dest_fact', 'plane_age'],
                                outputCol='features')

# Make the pipeline
flights_pipe = Pipeline(stages=[dest_indexer, dest_encoder, carr_indexer, carr_encoder, vec_assembler])

# Fit and transform the data
piped_data = flights_pipe.fit(model_data).transform(model_data)

# Split the data into training and test sets
training, test = piped_data.randomSplit([.6, .4])
```

```python
# Call lr.fit()
best_lr = lr.fit(training)

# Print best_lr
print(best_lr)
```

```python
# Use the model to predict the test set
test_results = best_lr.transform(test)

# Evaluate the predictions
print(evaluator.evaluate(test_results))
```
