###**Apprentissage automatique et Spark**

####**1. Apprentissage automatique et Spark**

+ ***Salut! Bienvenue dans le cours sur l’apprentissage automatique avec Apache Spark, dans lequel vous apprendrez à construire des modèles de machine learning sur de grands ensembles de données à l’aide de techniques de calcul distribué. Commençons par quelques concepts fondamentaux.***

####**2. Construire la gaufre parfaite (une analogie)**
+ **Supposons que vous vouliez apprendre à un ordinateur à faire des gaufres.**
+ **Vous pouvez trouver une bonne recette, puis donner à l’ordinateur des instructions explicites sur les ingrédients et les proportions.**
+ **Alternativement, vous pouvez présenter à l’ordinateur une sélection de différentes recettes de gaufres et le laisser déterminer les ingrédients et les proportions de la meilleure recette.**
+ ***La deuxième approche est le fonctionnement de l’apprentissage automatique : l’ordinateur apprend littéralement à partir d’exemples.***

####**3. Régression et classification**
+ Les problèmes d’apprentissage automatique sont généralement moins ésotériques que de trouver la recette de gaufres parfaite.
+ Les problèmes les plus courants concernent la régression ou la classification.
+ Un modèle de régression apprend à prédire un nombre.
+ Par exemple, lors de la fabrication de gaufres, quelle quantité de farine faut-il utiliser pour une quantité particulière de sucre?
+ Un modèle de classification, en revanche, prédit une valeur discrète ou catégorielle.
+ Par exemple, une recette qui demande une certaine quantité de sucre et de sel est-elle plus susceptible d’être pour les gaufres ou les cupcakes?

####**4. Données dans la RAM**
+ Les performances d’un modèle Machine Learning dépendent des données.
+ En général, plus de données est une bonne chose.
+ Si un algorithme est capable de s’entraîner sur un plus grand ensemble de données, sa capacité à généraliser à de nouvelles données s’améliorera inévitablement.
+ Cependant, il existe certaines contraintes pratiques.
+ Si les données peuvent tenir entièrement dans la RAM, l’algorithme peut fonctionner efficacement.
+ Que se passe-t-il lorsque ces données ne rentrent plus dans la mémoire ?

####**5. Les données dépassent la RAM**
+ L’ordinateur commencera à utiliser la * mémoire virtuelle * et les données seront * paginées * entre la RAM et le disque.
+ Par rapport à l’accès RAM, la récupération des données du disque est lente.

+ À mesure que la taille des données augmente, la pagination devient plus intense et l’ordinateur commence à passer de plus en plus de temps à attendre les données. Les performances s’effondrent.

####**6. Données distribuées sur un cluster**

+ **Comment traitons-nous alors des ensembles de données vraiment volumineux?**

+ **Une option consiste à répartir le problème sur plusieurs ordinateurs d’un cluster.**

+ **Plutôt que d’essayer de gérer un grand jeu de données sur une seule machine, il est divisé en partitions qui sont traitées séparément. Idéalement, chaque partition de données peut tenir dans la RAM sur un seul ordinateur du cluster. C’est l’approche utilisée par Spark.**

####**7. Qu’est-ce que Spark ?**
+ **Spark est un framework à usage général pour l’informatique en cluster. Il est populaire pour deux raisons principales**:
  + ***1. il est généralement beaucoup plus rapide que d’autres technologies Big Data comme Hadoop, car il effectue la plupart des traitements en mémoire**  
  + ***2. Il a une interface conviviale pour les développeurs qui cache une grande partie de la complexité de l’informatique distribuée.***

####**8. Composants: nœuds**
+**Passons en revue les composants d’un cluster Spark**.
+ **Le cluster lui-même se compose d’un ou plusieurs nœuds**.
+ **Chaque nœud est un ordinateur avec CPU, RAM et stockage physique**.

####**9. Composants: gestionnaire de cluster**
+ **Un gestionnaire de cluster alloue des ressources et coordonne l’activité dans l’ensemble du cluster.**

####**10. Composants: pilote**
+ Chaque application exécutée sur le cluster Spark possède un programme pilote.
+ **À l’aide de l’API Spark, le pilote communique avec le gestionnaire de cluster, qui à son tour distribue le travail aux nœuds.**

####**11. Composantes : exécuteurs testamentaires**
+ Sur chaque nœud, Spark lance un processus d’exécution qui persiste pendant toute la durée de l’application.

+ Le travail est divisé en tâches, qui sont simplement des unités de calcul.

+ Les exécuteurs exécutent des tâches dans plusieurs threads sur les cœurs d’un nœud.

+ Lorsque vous travaillez avec Spark, vous n’avez normalement pas à vous soucier * trop * des détails du cluster.

+ Spark configure toute cette infrastructure pour vous et gère toutes les interactions au sein du cluster. Cependant, il est toujours utile de savoir comment cela fonctionne sous le capot.

####**12. En avant !**
+ Vous avez maintenant une compréhension de base des principes de l’apprentissage automatique et de l’informatique distribuée avec Spark.

####**2. Anatomie d’un arbre de décision : nœud racine**

+ ***Un arbre de décision est construit à l’aide d’un algorithme appelé « partitionnement récursif »***.

+ **Prenons un exemple hypothétique dans lequel vous construisez un arbre de décision pour diviser les données en deux classes, verte et bleue**.

+ **Vous commencez par placer tous les enregistrements dans le nœud racine.**

+ Supposons qu’il y ait plus d’enregistrements verts que bleus, auquel cas ce nœud sera étiqueté « vert ».

+ Maintenant, parmi les prédicteurs dans les données, vous devez choisir celui qui se traduira par la division la plus informative des données en deux groupes.

+ Idéalement, vous voulez que les groupes soient aussi homogènes (ou « purs ») que possible: l’un devrait être principalement vert et l’autre devrait être principalement bleu.

#####**3. Anatomie d’un arbre de décision : première scission**

+ Une fois que vous avez identifié le prédicteur le plus informatif, vous divisez les données en deux ensembles, étiquetés « vert » ou « bleu » selon la classe dominante.

+ Et c’est là que la récursivité entre en jeu : vous appliquez alors exactement la même procédure sur chacun des nœuds enfants, en sélectionnant le prédicteur le plus informatif et en divisant à nouveau.

####**4. Anatomie d’un arbre de décision : deuxième division**

+ Ainsi, par exemple, le nœud vert à gauche pourrait être divisé à nouveau en deux groupes.

####**5. Anatomie d’un arbre de décision : troisième division**

+ Et le nœud vert résultant pourrait à nouveau être divisé. La profondeur de chaque branche de l’arbre n’a pas besoin d’être la même.

+ Il existe une variété de critères d’arrêt qui peuvent provoquer l’arrêt de la division le long d’une branche.

+ Par exemple, si le nombre d’enregistrements dans un nœud tombe en dessous d’un seuil ou si la pureté d’un nœud est supérieure à un seuil, vous pouvez arrêter le fractionnement.

+ Une fois que vous avez construit l’arbre de décision, vous pouvez l’utiliser pour faire des prédictions pour de nouvelles données en suivant les divisions du nœud racine jusqu’à la pointe d’une branche.

+ L’étiquette du nœud final serait alors la prédiction des nouvelles données.

+ ***`Une matrice de confusion` donne une ventilation utile des prédictions par rapport aux valeurs connues. Il comporte quatre cellules qui représentent les comptes de :***

  + ***`Vrais négatifs (TN)` - le modèle prédit un résultat négatif et le résultat connu est négatif***
  + ***`Vrais positifs (TP)` - le modèle prédit un résultat positif et le résultat connu est positif***
  + ***`Faux négatifs (FN)` - le modèle prédit un résultat négatif mais le résultat connu est positif***
  + ***`Faux positifs (FP)` - le modèle prédit un résultat positif mais le résultat connu est négatif.***


+ ***Ces décomptes (TN, TP, FN et FP) doivent totaliser le nombre d'enregistrements dans les données de test, qui ne constituent qu'un sous-ensemble des données de vols. Vous pouvez comparer le nombre d'enregistrements dans les données de tests, qui est flight_test.count().***

+ Calculez la précision.

```python
# Create a confusion matrix
prediction.groupBy('label', 'prediction').count().show()

# Calculate the elements of the confusion matrix
TN = prediction.filter('prediction = 0 AND label = prediction').count()
TP = prediction.filter('prediction = 1 AND label = prediction').count()
FN = prediction.filter('prediction = 0 AND label != prediction').count()
FP = prediction.filter('prediction = 1 AND label != prediction').count()

# Accuracy measures the proportion of correct predictions
accuracy = (TP+TN)/(TP+TN+FN+FP)
print(accuracy)
```

+ **Toutes ces mesures ont supposé un seuil de la moitié. Que se passe-t-il si vous modifiez ce seuil?**

####**14. ROC et AUC**
+ ***Un seuil est utilisé pour décider si le nombre renvoyé par le modèle de régression logistique se traduit par la classe positive ou négative**.

+ **Par défaut, ce seuil est fixé à la moitié. Cependant, ce n’est pas le seul choix.**

+ **Le choix d’une valeur supérieure ou inférieure pour le seuil affectera les performances du modèle**.

+ **La courbe ROC représente le taux de vrais positifs par rapport au taux de faux positifs lorsque le seuil passe de zéro (en haut à droite) à un (en bas à gauche)**.

+ **L’AUC résume la courbe ROC en un seul nombre.**

+ **C’est littéralement la zone sous la courbe ROC. L’AUC indique les performances d’un modèle sur toutes les valeurs du seuil**.

+ **Un modèle idéal, qui fonctionne parfaitement quel que soit le seuil, aurait une AUC de 1. Dans un exercice, nous verrons comment utiliser un autre évaluateur pour calculer l’AUC.**

###**Régularisation**

####**1. Régularisation**

+ ***Les modèles de régression que vous avez construits jusqu’à présent ont inclus aveuglément toutes les fonctionnalités fournies. Ensuite, vous allez en apprendre davantage sur un modèle plus sophistiqué qui sélectionne efficacement uniquement les fonctionnalités les plus utiles.***

#####**2. Caractéristiques: Seulement quelques-uns**


+ **Un modèle de régression linéaire tente de dériver un coefficient pour chaque caractéristique des données. Les coefficients quantifient l’effet des caractéristiques correspondantes. Plus de fonctionnalités impliquent plus de coefficients. Cela fonctionne bien lorsque votre jeu de données comporte quelques colonnes et de nombreuses lignes. Vous devez dériver quelques coefficients et vous avez beaucoup de données.***

####**3. Caractéristiques: Trop**
+ La situation inverse, de nombreuses colonnes et peu de lignes, est beaucoup plus difficile.

+ Maintenant, vous devez calculer des valeurs pour de nombreux coefficients, mais vous n’avez pas beaucoup de données pour le faire.

+ Même si vous parvenez à dériver des valeurs pour tous ces coefficients, votre modèle finira par être très compliqué et difficile à interpréter.

+ Idéalement, vous voulez créer un modèle parcimonieux : un modèle qui a juste le nombre minimum requis de prédicteurs.

+ Ce sera aussi simple que possible, tout en restant capable de faire des prédictions robustes.

####**4. Caractéristiques: Sélectionné**
+ ***La solution évidente consiste simplement à sélectionner le « meilleur » sous-ensemble de colonnes. Mais comment choisir ce sous-ensemble? Il existe une variété d’approches à ce problème de « sélection de fonctionnalités ».***

####**5. Fonction de perte (revisitée)**
+ **Dans cette leçon, nous explorerons l’une de ces approches de sélection de caractéristiques connue sous le nom de « régression pénalisée ».**

+ **L’idée de base est que le modèle est pénalisé, ou puni, pour avoir trop de coefficients.**

+ **Rappelons que l’algorithme de régression classique choisit des coefficients pour minimiser la fonction de perte, qui est la moyenne des résidus au carré. Un bon modèle se traduira par un MSE faible car ses prédictions seront proches des valeurs observées.**

####**6. Fonction de perte avec régularisation**
+ **Avec la régression pénalisée, un terme supplémentaire de « régularisation » ou de « rétrécissement » est ajouté à la fonction de perte. Plutôt que de dépendre des données, ce terme est fonction des coefficients du modèle.**

####**7. Délai de régularisation**

+ **Il existe deux formulaires standard pour le terme de régularisation**.

+ **La régression Lasso utilise un terme qui est proportionnel à la valeur absolue des coefficients, tandis que la régression de Ridge utilise le carré des coefficients**.

+ **Dans les deux cas, ce terme supplémentaire dans la fonction de perte pénalise les modèles avec trop de coefficients**.

+ **Il y a une distinction subtile entre la régression Lasso et Ridge**.

+ **Les deux réduiront les coefficients des prédicteurs sans importance**.

+ **Cependant, alors que Ridge fera en sorte que ces coefficients seront proches de zéro, Lasso les forcera en fait à zéro précisément**.

+  **Il est également possible d’avoir un mélange de Lasso et de Ridge**.

+ **La force de la régularisation est déterminée par un paramètre qui est généralement désigné par le symbole grec lambda**.

+  **Lorsque $λ = 0$ il n’y a pas de régularisation et lorsque lambda est grand, la régularisation domine complètement. Idéalement, vous voulez choisir une valeur pour lambda entre ces deux extrêmes!**

###**Pipeline**

####**1. Pipeline**
+ ***Bienvenue! Jusqu’à présent, vous avez appris à créer des modèles de classification et de régression à l’aide de Spark. Dans ce chapitre, vous apprendrez comment améliorer ces modèles. Vous commencerez par jeter un coup d’œil aux pipelines, ce qui rationalisera sérieusement votre flux de travail. Ils contribueront également à garantir que les données de formation et d’essai sont traitées de manière cohérente et qu’il n’y a pas de fuite d’informations entre ces deux ensembles.***

####**2. Fuite?**


+ **Qu’est-ce que j’entends par fuite?**

+ **La plupart des actions que vous avez utilisées impliquent à la fois une méthode fit() et une méthode transform().**

+ **Ces méthodes ont été appliquées de manière assez souple. Mais pour obtenir des résultats vraiment robustes, vous devez faire attention à appliquer la méthode fit() aux données d’entraînement**.

+ **Pourquoi? Parce que si une méthode fit() est appliquée à *n’importe la* des données de test, le modèle aura effectivement vu ces données pendant la phase d’apprentissage, de sorte que les résultats des tests ne seront plus objectifs**.

+ **La méthode transform(), en revanche, peut être appliquée à la fois aux données d’apprentissage et de test, car elle n’entraîne aucune modification du modèle sous-jacent.**

####**3. Un modèle qui fuit**
+ Un chiffre devrait rendre cela plus clair.

+ **Une fuite se produit chaque fois qu’une méthode fit() est appliquée aux données de test**.

+ **Supposons que vous adaptiez un modèle à l’aide des données d’apprentissage et de test**.

+ Le modèle aurait alors déjà *vu* les données de test, donc utiliser ces données pour tester le modèle ne serait pas juste :
  + bien sûr, le modèle fonctionnera bien sur les données qui ont été utilisées pour l’entraînement !
  + Cela semble évident, mais il faut veiller à ne pas tomber dans ce piège.
  + Rappelez-vous qu’il y a normalement plusieurs étapes dans la construction d’un modèle et si la méthode fit() dans *n’importe la* de ces étapes est appliquée aux données de test, alors le modèle est compromis.

####**4. Un modèle étanche**
+ Cependant, si vous prenez soin de n’appliquer fit() qu’aux données d’entraînement, votre modèle sera en bon état.

+ En ce qui concerne les tests, il n’aura vu *aucune* des données de test et les résultats des tests seront complètement objectifs.

+ **Heureusement, un pipeline facilitera l’évitement des fuites, car il simplifie le processus de formation et de test.**

####**5. Pipeline**
+ **Un pipeline est un mécanisme permettant de combiner une série d’étapes. Plutôt que d’appliquer chacune des étapes individuellement, elles sont toutes regroupées et appliquées comme une seule unité.**

####**6. Modèle de voitures: Étapes**
+ Revenons à notre modèle de régression des voitures. Rappelez-vous qu’il y avait un certain nombre d’étapes impliquées:
  - **l’utilisation d’un indexeur de chaînes pour convertir la colonne type en valeurs indexées;**
  - **l’application d’un codeur à chaud unique pour convertir ces valeurs indexées en variables fictives; alors **
  - **l’assemblage d’un ensemble de prédicteurs en une seule colonne de caractéristiques; et enfin**
  - **construction d’un modèle de régression.**

####**7. Modèle de voitures: application des étapes**
+ **Cartographions le processus d’application de ces étapes**.
  - **Tout d’abord, vous ajustez l’indexeur aux données d’apprentissage.**
  + **Ensuite, vous appelez la méthode transform() sur les données d’apprentissage pour ajouter la colonne indexée.**
  - **Ensuite, vous appelez la méthode transform() sur les données de test pour y ajouter la colonne indexée. Notez que les données de test n’ont pas été utilisées pour s’adapter à l’indexeur.**
  + **Ensuite, vous faites la même chose pour le codeur one-hot, en l’ajustant aux données d’apprentissage, puis en utilisant le codeur ajusté pour mettre à jour les ensembles de données d’apprentissage et de test.**
  + **L’assembleur est le suivant. Dans ce cas, il n’y a pas de méthode fit(), vous appliquez donc simplement la méthode transform() aux données d’apprentissage et de test. Enfin, les données sont prêtes**.
  + **Vous ajustez le modèle de régression aux données d’apprentissage, puis utilisez le modèle pour faire des prédictions sur les données de test**.
  + **Tout au long du processus, vous avez pris soin de garder les données de test en dehors du processus de formation. Mais c’est un travail difficile et il est assez facile de glisser.**

####**8. Modèle de voitures: Pipeline**
+ ***Un pipeline facilite grandement la formation et le test d’un modèle compliqué. La classe Pipeline réside dans le sous-module ml. Vous créez un pipeline en spécifiant une séquence d’étapes, où chaque étape correspond à une étape du processus de création du modèle. Les étapes sont exécutées dans l’ordre. Maintenant, plutôt que d’appeler les méthodes fit() et transform() pour chaque étape, vous appelez simplement la méthode fit() pour le pipeline sur les données d’apprentissage. Chacune des étapes du pipeline est ensuite automatiquement appliquée aux données d’entraînement à tour de rôle. Cela appliquera systématiquement les méthodes fit() et transform() pour chaque étape du pipeline. Le pipeline entraîné peut ensuite être utilisé pour faire des prédictions sur les données de test en appelant sa méthode transform(). La méthode transform() du pipeline appelle uniquement la méthode transform() pour chacune des étapes du pipeline. N’est-ce pas simple?***

####**9. Modèle de voitures: Étapes**

+ **Vous pouvez accéder aux étapes du pipeline à l’aide de l’attribut .stages, qui est une liste. Vous choisissez des étapes individuelles en les indexant dans la liste. Par exemple, pour accéder au composant de régression du pipeline, vous devez utiliser un index de 3. L’accès à ce composant permet d’obtenir l’interception et les coefficients pour le modèle de régression linéaire formé.**

###**Validation croisée**

####**1. Validation croisée**
+ ***Jusqu’à présent, vous testiez des modèles à l’aide d’une technique assez simple : diviser au hasard les données en ensembles d’entraînement et de test, entraîner le modèle sur les données d’entraînement, puis évaluer ses performances sur l’ensemble de test. Il y a un inconvénient majeur à cette approche : vous n’obtenez qu’une seule estimation des performances du modèle. Vous auriez une idée plus solide du fonctionnement d’un modèle si vous pouviez le tester plusieurs fois. C’est précisément l’idée derrière la validation croisée.***

####**2. CV - données complètes**
+ **Vous commencez avec l’ensemble complet des données.**

####**3. CV - séparation train/test**
+ **Vous divisez toujours ces données en un ensemble d’entraînement et un ensemble de tests. N’oubliez pas qu’avant de fractionner, il est important de randomiser d’abord les données afin que les distributions dans les données d’entraînement et de test soient similaires.**

####**4. CV - plis multiples**
+ **Vous divisez ensuite les données d’entraînement en un certain nombre de partitions ou de « plis ». Le nombre de plis entre normalement en ligne de compte dans le nom de la technique. Par exemple, si vous divisez en cinq volets, vous parlerez de validation croisée 5 fois.**

####**5. Pli sur pli - premier pli**
+ Une fois que les données d’entraînement ont été divisées en plis, vous pouvez commencer la validation croisée.
+ Gardez d’abord de côté les données dans le premier pli.
+ Entraînez un modèle sur les quatre plis restants.
+ Ensuite, évaluez ce modèle sur les données du premier pli.
+ Cela donnera la première valeur de la métrique d’évaluation.

####**6. Pli sur pli - deuxième pli**
+ Ensuite, vous passez au deuxième pli, où le même processus est répété:
  + les données du deuxième pli sont mises de côté pour les tests tandis que les quatre plis restants sont utilisés pour former un modèle.
  + Ce modèle est testé sur les données du deuxième pli, ce qui donne la deuxième valeur pour la mesure d’évaluation.

####**7. Pli sur pli - autres plis**
+ Vous répétez le processus pour les plis restants.
+ Chacun des plis est utilisé à tour de rôle comme données de test et vous obtenez autant de valeurs pour la métrique d’évaluation qu’il y a de plis.
+ À ce stade, vous êtes en mesure de calculer la moyenne de la métrique d’évaluation sur tous les plis, ce qui est une mesure beaucoup plus robuste des performances du modèle qu’une valeur unique.

####**8. Les voitures revisitées**
+ Voyons comment cela fonctionne dans la pratique.
+ Vous vous souvenez des données des voitures ? Bien sûr que oui.
+ Vous allez construire un modèle de régression à validation croisée pour prédire la consommation.

####**9. Estimateur et évaluateur**
+ Voici les deux premiers ingrédients dont vous avez besoin pour effectuer une validation croisée :
  - un estimateur, qui construit le modèle et est souvent un pipeline ; et
  - un évaluateur, qui quantifie le fonctionnement d’un modèle sur les données de test.
  + Nous avons déjà vu les deux à quelques reprises.

####**10. Grille et validateur croisé**
+ **Maintenant, les ingrédients finaux.**
+ **Vous aurez besoin de deux nouvelles classes, CrossValidator et ParamGridBuilder, toutes deux issues du sous-module de réglage**.
+ **Vous allez créer une grille de paramètres, que vous laisserez vide pour le moment, mais sur laquelle vous reviendrez en détail lors de la prochaine leçon**.

+ Enfin, vous avez tout ce qu’il faut pour créer un objet validateur croisé :
  - **un estimateur, qui est le modèle de régression linéaire**,
  - **une grille vide de paramètres pour l’estimateur et**
  - **un évaluateur qui calculera le RMSE**.
  + **Vous pouvez éventuellement spécifier le nombre de plis (qui est par défaut trois) et une amorce de nombres aléatoires pour la répétabilité.**

####**11. Les validateurs croisés ont aussi besoin de formation**
+ **Le validateur croisé dispose d’une méthode fit() qui appliquera la procédure de validation croisée aux données d’apprentissage.**
+ **Vous pouvez ensuite regarder le RMSE moyen calculé sur tous les plis.**
+ **Il s’agit d’une mesure plus robuste de la performance du modèle, car elle est basée sur plusieurs fractionnements train/test.**
+ **Notez que la métrique moyenne est renvoyée sous forme de liste. Vous verrez pourquoi dans la prochaine leçon.**

####**12. Les validateurs croisés agissent comme des modèles**
+ L’objet validateur croisé entraîné agit comme n’importe quel autre modèle.
+ Il a une méthode de transformation, qui peut être utilisée pour faire des prédictions sur de nouvelles données.
+ Si nous évaluons les prédictions sur les données de test originales, nous trouvons une valeur plus faible pour le RMSE que celle obtenue par validation croisée.
+ Cela signifie qu’une simple séparation train-test aurait donné une vision trop optimiste des performances du modèle.

####**2. Qu’est-ce qu’un ensemble?**

+ **En termes simples, un modèle d’ensemble n’est qu’une collection de modèles. Un modèle d’ensemble combine les résultats de plusieurs modèles pour produire de meilleures prédictions que n’importe lequel de ces modèles agissant seul. Le concept est basé sur l’idée de la « sagesse de la foule », ce qui implique que l’opinion agrégée d’un groupe est meilleure que les opinions des individus de ce groupe, même si les individus sont des experts.**

####**3. Diversité d’ensemble**

+ **Comme le suggère la citation, pour que cette idée soit vraie, il doit y avoir de la diversité et de l’indépendance dans la foule. Cela s’applique également aux modèles : un ensemble réussi nécessite des modèles divers. Cela n’aide pas si tous les modèles de l’ensemble sont similaires ou exactement les mêmes. Idéalement, chacun des modèles de l’ensemble devrait être différent.**

####**4. Forêt aléatoire**
+ **Une forêt aléatoire, comme son nom l’indique, est une collection d’arbres. Pour s’assurer que chacun de ces arbres est différent, l’algorithme de l’arbre de décision est légèrement modifié : - chaque arbre est formé sur un sous-ensemble aléatoire différent des données et - Dans chaque arbre, un sous-ensemble aléatoire de caractéristiques est utilisé pour la division à chaque nœud. Le résultat est une collection d’arbres où il n’y a pas deux arbres identiques. Dans le modèle de forêt aléatoire, tous les arbres fonctionnent en parallèle.**

+ **Vous trouverez de bonnes valeurs pour les paramètres suivants :**

  + featureSubsetStrategy - le nombre de fonctionnalités à prendre en compte pour le fractionnement à chaque nœud et
  + maxDepth — le nombre maximum de divisions le long de n'importe quelle branche.

+ *Malheureusement, la construction de ce modèle prend trop de temps, nous n'exécuterons donc pas la méthode .fit() sur le pipeline.*

####**Instructions**

+ Créez un objet classificateur de forêt aléatoire.
+ Créez un objet générateur de grille de paramètres.
+ Ajoutez des points de grille pour les paramètres featureSubsetStrategy et maxDepth.
+ Créer un évaluateur de classification binaire.
+ Créez un objet de validation croisée, en spécifiant l'estimateur, la grille de paramètres et l'évaluateur.
+ Choisissez la validation croisée quintuple.

```python
# Create a random forest classifier
forest = RandomForestClassifier()

# Create a parameter grid
params = ParamGridBuilder() \
            .addGrid(forest.featureSubsetStrategy, ['all', 'onethird', 'sqrt', 'log2']) \
            .addGrid(forest.maxDepth, [2, 5, 10]) \
            .build()

# Create a binary classification evaluator
evaluator = BinaryClassificationEvaluator()

# Create a cross-validator
cv = CrossValidator(estimator=forest, estimatorParamMaps=params, evaluator=evaluator, numFolds=5)
```

***Excellent! Une recherche de grille peut être utilisée pour optimiser tous les paramètres d'un pipeline de modèle.***
