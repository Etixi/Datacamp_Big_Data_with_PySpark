###**Introduction à PySpark RDD**

####**1. Introduction à PySpark RDD**

+ ***Dans le premier chapitre, vous avez découvert différents composants de Spark, à savoir Spark Core, Spark SQL et Spark MLlib.***

+ *Dans ce chapitre, nous commencerons par* ***les RDD qui sont l’abstraction de base de Spark pour travailler avec des données***.

####**2. Qu’est-ce que RDD?**

+ *Commençons*. ***RDD signifie Resilient Distributed Datasets***.

+ ***Il s’agit simplement d’une collection de données réparties dans le cluster***.

+ ***RDD est le type de données fondamental et de base dans PySpark***.

+ ***Lorsque Spark commence à traiter des données, il divise les données en partitions et distribue les données sur les nœuds de cluster, chaque nœud contenant une tranche de données***.

####**3. Décomposition des RDD**

+ Maintenant, prenons un regardez les différentes fonctionnalités de RDD.

+ Le nom RDD capture 3 propriétés importantes.

+ ***Résilient ou Resilient, ce qui signifie la capacité de résister aux défaillances et de recalculer les partitions manquantes ou endommagées***.

+ ***Distribué ou Distributed, ce qui signifie répartir les tâches sur plusieurs nœuds du cluster pour un calcul efficace***.

+ ***Datasets, qui est une collection de données partitionnées, par exemple des tableaux, des tables, des tuples ou d’autres objets***.

###**Introduction aux DataFrames PySpark**

####**1.Introduction aux DataFrames PySpark**

+ *Dans le chapitre précédent, vous avez examiné les RDD qui sont l’abstraction de base de Spark pour travailler avec des données. Dans ce chapitre, nous explorerons PySpark SQL qui est l’API de haut niveau de Spark pour travailler avec des données structurées.*

####**2. Que sont les DataFrames PySpark ?**
+ ***PySpark SQL est une bibliothèque Spark pour les données structurées.***

+ Contrairement à l’API **RDD PySpark, PySpark SQL** fournit plus d’informations sur la structure des données et le calcul effectué.

+ **PySpark SQL fournit une abstraction de programmation appelée DataFrames**.

+ **Un DataFrame est une collection distribuée immuable de données avec des colonnes nommées**.

+ Il est similaire à une table en SQL.

+ ***Les DataFrames sont conçus pour traiter une grande collection de données structurées telles que des bases de données relationnelles et des données semi-structurées telles que JSON (JavaScript Object Notation)***.

+ L’API DataFrame prend actuellement en charge plusieurs langages tels que ***Python, R, Scala et Java. DataFrames permet à PySpark d’interroger des données en utilisant SQL, par exemple (SELECT * from table) ou en utilisant la méthode d’expression par exemple (df-dot-select)***.

####**3. SparkSession - Point d’entrée pour l’API DataFrame**

+ Auparavant, vous avez appris à propos de SparkContext qui est le point d’entrée principal pour la création de RDD.

+ De même, SparkSession fournit un point d’entrée unique pour interagir avec la fonctionnalité Spark sous-jacente et permet de programmer Spark avec l’API DataFrame.

+ Le ***SparkSession fait pour DataFrames ce que le SparkContext fait pour les RDD***.

+ ***Un SparkSession peut être utilisé pour créer DataFrame, enregistrer DataFrame en tant que tables, exécuter SQL sur des tables, mettre en cache des tables, etc., Semblable à SparkContext, SparkSession est exposé au shell PySpark en tant qu’étincelle variable***.

####**4. Création de DataFrames dans PySpark**

+ DataFrames dans Pyspark peut être créé de deux manières principales.

+ À partir d’un RDD existant à l’aide de la méthode createDataFrame de SparkSession et À partir de différentes sources de données telles que CSV, JSON, TXT en utilisant la méthode de lecture de SparkSession. Avant d’entrer dans les détails de la création de DataFrames, comprenons ce qu’est le schéma.

+ Schema est la structure des données dans DataFrame et aide Spark à optimiser les requêtes sur les données plus efficacement.

+ Un schéma fournit des informations telles que le nom de la colonne, le type de données dans cette colonne et si les valeurs null ou vides sont autorisées dans la colonne.
